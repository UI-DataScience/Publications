```python
import matplotlib.pyplot as plt

plt.rcParams.update({'font.size': 22})


import numpy as np
import seaborn as sns
```

```python
data_size = [1000, 10000, 100000, 1000000]
x = [i for i in range(0, len(data_size))]

# Wall-Time on Nebula
'''
IDTxl_wall_time = [16.2, 105, 1038, 25171 ]
TransEnt_wall_time = [0.02, 0.371, 25.653, 2304.763]
RTransferEntropy_wall_time = [4.26, 13.817, 94.889, 622.369]
LCDM_wall_time = [2.32, 1.32, 3.04, 129]
LCDM_wall_time_GPU = [] 
'''

# Wall-Time on Nano. Whole node 6; 2 GPUS
# qsub -I -l nodes=nano6:ppn=12:gpus=2,walltime=14400
LCDM_wall_time = [2.564, 3.138, 6.820, 3*60+57.122]
LCDM_wall_time_GPU =[3.282, 1.980, 3.996, 40.231]

TransEnt_wall_time = [0.656, 0.968, 21.708, 26 * 60 + 25.942]


IDTxl_wall_time = [10.980, 60*1 + 40.230, 60 * 17 + 50.749, 719 * 60 + 10.129]
RTransferEntropy_wall_time = [12.492, 60*1+42.228, 60 * 17 + 16.661, 176 *60 + 32.770]

RTransferEntropy_wall_time_parallel = [2.876, 15.703, 2 * 60 + 7.281, 60*19 + 48.636 ]

TRENTTOOL3_wall_time = []
TRENTTOOL3_wall_time_GPU = []
```

```python
color_pal = sns.color_palette(palette='colorblind')
```

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, np.log(IDTxl_wall_time), markersize=15, label="IDTxl", linestyle='--', marker='p', color=color_pal[6], alpha=.6)
ax.plot(x, np.log(TransEnt_wall_time),markersize=10, label="TransEnt", marker='v', color=color_pal[1], alpha=.6)
ax.plot(x, np.log(RTransferEntropy_wall_time), label="RTransferEntropy", linestyle='-.', marker='s', color=color_pal[2], alpha=.6)
ax.plot(x, np.log(RTransferEntropy_wall_time_parallel),markersize=10, label="RTransferEntropy (Parallel)", marker='1', color=color_pal[3], alpha=.6)
ax.plot(x, np.log(LCDM_wall_time),markersize=10, label="PyIF (CPU)", marker='*', color=color_pal[4], alpha=.6)
ax.plot(x, np.log(LCDM_wall_time_GPU),markersize=10, label="PyIF (GPU)", marker='D', color=color_pal[5], alpha=.6)



ax.set_xlabel('Number of Random Samples')
ax.set_ylabel('Natural Log Seconds')

ax.set_xticks([0, 1, 2, 3])
ax.set_xticklabels(["1,000", "10,000", "100,000", "1,000,000"])

ax.legend(prop={"size":16.5})


plt.show()

fig.savefig('WallTime-TE.png')
```
